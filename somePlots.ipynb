```python
import numpy as np
im = np.load('Data/ImagesCreated/image_40x40x25_1.npy')
imDim = im.shape
imDim
```

```python
X,Y,Z = [],[],[]
for x in range(imDim[0]):
    for y in range(imDim[1]):
        for z in range(imDim[2]):
            if im[x,y,z] == 1:
                X.append(x)
                Y.append(y)
                Z.append(z)
```

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(X,Y,Z, color='#800026',s=3)
ax.plot3D([0,40,40,0,0],[0,0,40,40,0],[0,0,0,0,0],'k')
ax.plot3D([0,40,40,0,0],[0,0,40,40,0],[25,25,25,25,25],'k')
ax.plot3D([0,0],[0,0],[0,25],'k')
ax.plot3D([0,0],[40,40],[0,25],'k')
ax.plot3D([40,40],[0,0],[0,25],'k')
ax.plot3D([40,40],[40,40],[0,25],'k')
ax.xaxis.set_pane_color('#ffffcc')
ax.yaxis.set_pane_color('#ffffcc')
ax.zaxis.set_pane_color('#ffffcc')
ax.xaxis.line.set_color('#ffffcc')
ax.yaxis.line.set_color('#ffffcc')
ax.zaxis.line.set_color('#ffffcc')
ax.axes.get_xaxis().set_ticks([])
ax.axes.get_yaxis().set_ticks([])
ax.axes.get_zaxis().set_ticks([])
ax.grid(False)
fig.show()
```
